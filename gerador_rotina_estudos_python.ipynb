```python
import matplotlib.pyplot as plt


print("GERADOR INTELIGENTE DE ROTINA DE ESTUDOS")

materias = []
total_tempo = float(input("Quanto tempo você tem para estudar por dia (em horas)? "))

while True:
    nome = input("\nNome da matéria (ou 'sair' para finalizar): ").title()
    if nome.lower() == "sair":
        break

    dificuldade = int(input("Dificuldade (1=fácil, 5=muito difícil): "))
    prioridade = int(input("Prioridade (1=baixa, 5=alta): "))

    materias.append({
        "nome": nome,
        "dificuldade": dificuldade,
        "prioridade": prioridade
    })

# Cálculo de peso por matéria
peso_total = 0
for m in materias:
    m["peso"] = m["dificuldade"] + m["prioridade"]
    peso_total += m["peso"]

print("\nROTINA DE ESTUDOS SUGERIDA")
for m in materias:
    proporcao = m["peso"] / peso_total
    tempo_materia = proporcao * total_tempo

    print(f"- {m['nome']}: {tempo_materia:.2f} horas por dia")

print("\nO plano foi criado com base em dificuldade + prioridade.")

labels = [m["nome"] for m in materias]
tempos = [(m["peso"] / peso_total) * total_tempo for m in materias]

plt.figure(figsize=(9, 9))

# Gráfico
patches, texts, autotexts = plt.pie(
    tempos,
    labels=None,              # removemos os labels diretos da pizza
    autopct='%1.1f%%',
    startangle=90
)

# Legenda ao lado
plt.legend(
    patches,
    labels,
    title="Matérias",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Distribuição do tempo de estudo por matéria")
plt.tight_layout()
plt.show()
```
